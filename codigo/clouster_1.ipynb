```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df_data = pd.read_csv('./data_con_articulos.csv')
```

```python
df_data.head(1)
```

```python
df_data_2022 = df_data[df_data.fecha.between('2022-01-01', '2023-01-01')]
```

```python
df_data_2022.head(1)
```

```python
df_data_2022.columns[:18]
```

```python
columnas_drop = ['creado', 'fecha']
```

```python
df_data_2022.drop(columns=columnas_drop,inplace=True)
```

```python
df_data_cliente_2022 = df_data_2022.groupby('cliente').sum()
```

```python
df_data_2022.groupby('cliente').proyecto.nunique()
```

```python
df_data_cliente_2022
```

```python
df_data_cliente_2022['compras_cliente'] =  df_data_2022.groupby('cliente').proyecto.nunique()
```

```python
df_data_cliente_2022.head(1)
```

# escalamos

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_data_cliente_2022.describe()
```

```python
columnas = list(df_data_cliente_2022.describe().columns)
```

```python
df_data_min = df_data_cliente_2022[['cantidad',
 'costo',
 'impresión',
 'precio',
 'total',
 'mermas',
 'paqueteria_o_gasto_extra',
 'utilidad_real',
 'inversion',
 'utilidad_real/cantidad','compras_cliente']]
```

```python
df_data_min
```

```python
df_data_min_escalado = scaler.fit_transform(df_data_min)
```

```python
df_data_cliente_2022_escalado = scaler.fit_transform(df_data_cliente_2022)
```

```python
df_data_cliente_2022_escalado
```

```python
df_data_escalado = pd.DataFrame(df_data_cliente_2022_escalado, columns = columnas)
```

```python
df_data_escalado.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans4 = KMeans(n_clusters = 6).fit(df_data_escalado)
```

```python
K = range(1, 30)

inertias = []

for k in K:
    modelo = KMeans(n_clusters = k).fit(df_data_escalado)
    inertias.append(modelo.inertia_)
```

```python
plt.plot(K, inertias, marker = 'x')
plt.xticks(K)
plt.grid()
plt.show()
```

```python
K = range(1, 30)

inertias = []

for k in K:
    modelo = KMeans(n_clusters = k).fit(df_data_min_escalado)
    inertias.append(modelo.inertia_)
```

```python
plt.plot(K, inertias, marker = 'x')
plt.xticks(K)
plt.grid()
plt.show()
```

```python
kmeans4 = KMeans(n_clusters = 5).fit(df_data_min_escalado)
```

```python
columnas_2 = ['cantidad',
 'costo',
 'impresión',
 'precio',
 'total',
 'mermas',
 'paqueteria_o_gasto_extra',
 'utilidad_real',
 'inversion',
 'utilidad_real/cantidad','compras_cliente']
```

```python
df_data_min_escalado = pd.DataFrame(df_data_min_escalado, columns = columnas_2)
```
