```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df_data = pd.read_csv('./data_total_cluster_articulo_final_5.csv')
```

```python
df_data.head(1)
```

```python
df_data_2022 = df_data[df_data.fecha.between('2022-01-01', '2023-01-01')]
df_data_2019 = df_data[df_data.fecha.between('2019-01-01', '2020-01-01')]
df_data_2021 = df_data[df_data.fecha.between('2021-01-01', '2022-01-01')]
```

```python
df_data_19_21_22 = pd.concat([df_data_2019,df_data_2019, df_data_2021], ignore_index=True)
```

```python
df_data_19_21_22.columns[-20:]
```

```python
df_fecha_articulo = df_data_19_21_22.drop(columns=['creado', 'proyecto', 'vendedor', 'tipo_de_cliente', 'cliente',
                       'cantidad', 'articulo', 'costo', 'impresión', 'precio', 'total',
                       'mermas', 'paqueteria_o_gasto_extra', 'utilidad_real', 'inversion',
                       'utilidad_real/cantidad', 'cluster_names', 'articulo_unico'])
```

```python
df_fecha_articulo['fecha']=pd.to_datetime(df_fecha_articulo.fecha) 
```

```python
df_fecha_articulo_tabla = df_fecha_articulo.groupby(df_fecha_articulo.fecha.dt.month).sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_fecha_articulo_tabla_escalado = scaler.fit_transform(df_fecha_articulo_tabla)
```

```python
df_fecha_articulo_tabla_escalado.head(1)
```

```python
lista_mes=['enero','febrero','marzo','abril','mayo','junio','julio','agosto','septiembre','octubre','noviembre','diciembre']
```

```python
df_fecha_articulo_tabla_escalado.
```

```python
df_fecha_articulo_tabla_escalado = pd.DataFrame(df_fecha_articulo_tabla_escalado , columns=df_fecha_articulo_tabla.columns)
```

```python
df = df_fecha_articulo_tabla_escalado.T
```

```python
df.columns = lista_mes
```

```python
df_fecha_articulo_tabla_escalado = df.T
```

```python
from scipy.spatial.distance import pdist, squareform
```

```python
distancias_mes_19_21_22 = pd.DataFrame(1/(1 + squareform(pdist(df_fecha_articulo_tabla_escalado))), columns = df_fecha_articulo_tabla_escalado.index, index = df_fecha_articulo_tabla_escalado.index)
```

```python
distancias_mes_19_21_22
```

```python
def recomendaciones_cliente(cliente):
    distancias_cliente = distancias_mes_19_21_22[cliente].drop(index = [cliente])
    df_resto = df_fecha_articulo_tabla_escalado.T.drop(columns = [cliente])
    for otro_cliente, peso in distancias_cliente.items():
        df_resto[otro_cliente] = df_resto[otro_cliente] * peso

    return df_resto.sum(axis=1)
```

```python
recomendaciones_cliente('diciembre').sort_values(ascending=False).head(20)
```

```python
x=0
for cliente in lista_mes:
    data_articulo = recomendaciones_cliente(cliente).sort_values(ascending=False)
    data_articulo = data_articulo.to_frame().reset_index()
    
    if x>0:
        data_articulo_total = data_articulo_total.merge(data_articulo, on= 'index', how= 'inner')
        
    if x==0:
        data_articulo_total = data_articulo.copy()
        x=x+1
    
```

```python
data_articulo_total
```

```python
data_articulo_total.columns = ['articulo', 'enero','febrero','marzo','abril','mayo','junio','julio','agosto','septiembre','octubre','noviembre','diciembre']
```
