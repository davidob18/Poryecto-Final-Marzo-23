```python
class QuitarColumnas:
    def fit(self, X,y):
        return self
    def transform(self, X):
        return X[['creado', 'fecha', 'proyecto', 'vendedor', 'tipo_de_cliente', 'cliente',
                  'cantidad', 'articulo', 'costo', 'impresión', 'precio', 'total',
                  'mermas', 'paqueteria_o_gasto_extra', 'utilidad_real']]
```

```python
pipeline_vendedores = Pipeline([('remove_cols', QuitarColumnas()),
                                   ('mean_imputation', mdi.MeanMedianImputer(imputation_method = 'median', 
                                                                             variables = ['mermas'])),
                                   ('zero_imputation', mdi.ArbitraryNumberImputer(arbitrary_number=0,
                                                                                  variables = ['paqueteria_o_gasto_extra','impresión'])),
                                ('mode_imputation', mdi.CategoricalImputer(imputation_method = 'missing', 
                                                                           variables = [ 'cliente', 'articulo']))
                                 
                                    ], verbose = True)
```

```python
pipeline_vendedores.fit(df_compras_christian_copy)
```

```python
data = pipeline_vendedores.transform(df_compras_christian_copy)
```
